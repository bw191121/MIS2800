```python
import sqlite3
```

```python
conn = sqlite3.connect('HW3_AFC_North_APPG.db')
```

# Create A Strict Table

```python
conn.execute('''
CREATE TABLE HW_AFC_North_APPG(
  team_name TEXT PRIMARY KEY NOT NULL,
  coach TEXT,
  wins INTEGER,
  loses INTEGER,
  ties INTEGER)
  STRICT;
  ''')
```

# Add Team Stats

```python
conn.execute('''
INSERT INTO HW_AFC_North_APPG (team_name, coach, wins, loses, ties)
VALUES ('Steelers', 'Mike Tomlin', 10, 7, 0),
('Ravens', 'John Harbaugh', 8, 9, 0),
('Bengals', 'Zac Taylor', 6, 11, 0),
('Browns', 'Kevin Stefanski', 5, 12, 0);
''')
```

# Filtering Teams

```python
#select all the data where teams had more than 7 wins
query = "SELECT * FROM HW_AFC_North_APPG WHERE wins > 7"
results = conn.execute(query)
for row in results:
  print(row)
```

# Sorting Data

```python
#Select only the team_name and wins. Sort numercially from highest to lowest based on wins
query = "SELECT team_name, wins FROM HW_AFC_North_APPG ORDER BY wins DESC"
for row in conn.execute(query):
  print(row)
```
